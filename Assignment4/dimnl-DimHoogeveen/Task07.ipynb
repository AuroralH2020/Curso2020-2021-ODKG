Read the RDF file

```
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS
g = Graph()
g.namespace_manager.bind('ns', Namespace("http://somewhere#"), override=False)
g.namespace_manager.bind('vcard', Namespace("http://www.w3.org/2001/vcard-rdf/3.0#"), override=False)
g.parse(github_storage+"/resources/example6.rdf", format="xml")
```

**TASK 7.1: List all subclasses of "Person" with RDFLib and SPARQL**

```
# Set namespace 
ns = Namespace("http://somewhere#")
```

```
#RDFLib
from rdflib import paths # for property path

for subject in g.subjects(RDFS.subClassOf*paths.OneOrMore, ns.Person):
    print(subject)
```

```
#SPARQL

query71 = g.query("""  
    SELECT ?subclass
    WHERE { 
    ?subclass rdfs:subClassOf/rdfs:subClassOf* ns:Person . 
  }
  """,
  initNs = { "ns": ns}
)

for row in query71:
    print("%s" % row)
```

**TASK 7.2: List all individuals of "Person" with RDFLib and SPARQL (remember the subClasses)**


```
#RDFLib

for subject in g.subjects(paths.path_sequence(RDF.type, RDFS.subClassOf*paths.ZeroOrMore), ns.Person):
    print(subject)
```

```
#SPARQL

query72 = g.query("""  
    SELECT ?individual
    WHERE { 
    ?individual rdf:type/rdfs:subClassOf* ns:Person . 
  }
  """,
  initNs = { "ns": ns}
)

for row in query72:
    print("%s" % row)
```

**TASK 7.3: List all individuals of "Person" and all their properties including their class with RDFLib and SPARQL**


```
#RDFLib

for subject in g.subjects(paths.path_sequence(RDF.type, RDFS.subClassOf*paths.ZeroOrMore), ns.Person):
    for predicate, object  in g.predicate_objects(subject):
        print(subject, "with property:", predicate)
```

```
#SPARQL

query73 = g.query("""  
    SELECT ?individual ?property
    WHERE { 
    ?individual rdf:type/rdfs:subClassOf* ns:Person . 
    ?individual ?property ?object .
  }
  """,
  initNs = { "ns": ns}
)

for row in query73:
    print("%s with property: %s" % row)
```
